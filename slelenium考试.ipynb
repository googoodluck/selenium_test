```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
import time 
import platform

# 获取所有产品及对应的价格

# 启动Chrome浏览器
driver = webdriver.Chrome()

# 打开网页
website_URL = "http://10.113.178.219/page/2"
driver.get(website_URL)

# 使用XPath定位所有产品名称和价格元素
name_elements = driver.find_elements(By.XPATH,"//div[@class='card-title']/strong")
price_elements = driver.find_elements(By.XPATH,"//h3[@class='card-text']")
# print(name_elements,price_elements)

# 创建一个空列表存储产品名称和价格
products = []

# 遍历产品名称和价格元素，将其匹配成元组并存储到列表中
for name_element, price_element in zip(name_elements, price_elements):
    product_name = name_element.text
    product_price = price_element.text
    products.append((product_name, product_price))

# 打印匹配的产品名称和价格
for product in products:
    print("产品名称:", product[0])
    print("产品价格:", product[1])
    print()  # 打印空行分隔每个产品

# 关闭浏览器
driver.quit()
```

```python
# 遍历产品列表，根据关键词进行分类
classified_products = { 'ipad': [], 'air': [], 'bud': [],'apple': [],'huawei': [],}

for product_info in products:
    product_name = product_info[0].strip().lower()
    product_price = product_info[1]
    
    if "ipad" in product_name:
        classified_products['ipad'].append((product_name, product_price))
    if "air" in product_name:
        classified_products['air'].append((product_name, product_price))
    if "bud" in product_name:
        classified_products['bud'].append((product_name, product_price))
    if "apple" in product_name:
        classified_products['apple'].append((product_name, product_price))
    if "huawei" in product_name:
        classified_products['huawei'].append((product_name, product_price))

# 打印分类结果
for category, products in classified_products.items():
    print(category + "类:")
    for product_name, product_price in products:
        print(f"产品名称: {product_name}")
        print(f"产品价格: {product_price}")
    print()

# 关闭浏览器
driver.quit()
```

```python

# 获取每个apple产品的链接

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.common.action_chains import ActionChains

# 启动浏览器
driver = webdriver.Chrome()
clicked_links = []
# 保存初始窗口句柄
initial_window = driver.current_window_handle

for url in all_links:
    driver.get(url)
    
    # 使用XPath定位所有产品名称和价格元素
    name_elements = driver.find_elements(By.XPATH, "//div[@class='card-title']/strong")
    
    for i in name_elements:
        if i.text in apple_list:
            print(i.text)
            # 在新标签页中打开链接
            ActionChains(driver).key_down(Keys.CONTROL).click(i).key_up(Keys.CONTROL).perform()
            
            # 切换到新打开的标签页
            driver.switch_to.window(driver.window_handles[-1])
            current_url = driver.current_url
            print(current_url)
            clicked_links.append(current_url)
            
            # 在新标签页完成操作后，关闭该标签页并回到初始窗口
            driver.close()
            driver.switch_to.window(initial_window)

# # 获取点击后的当前页面链接并添加到列表

print(clicked_links)
```
